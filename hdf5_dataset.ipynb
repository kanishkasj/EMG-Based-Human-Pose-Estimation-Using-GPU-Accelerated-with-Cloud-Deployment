```python
import h5py
import pandas as pd
import numpy as np
import os
from scipy.signal import butter, filtfilt
from google.colab import files # Import files module

# ==== 0. Upload HDF5 files ====
print("Please upload the required HDF5 files now:")

# ==== 1. Load metadata.csv ====
metadata_path = "/content/preprocess_metadata.csv"
if not os.path.exists(metadata_path):
    print(f"Error: Metadata file not found at {metadata_path}")
else:
    meta = pd.read_csv(metadata_path)
    print(f"Loaded metadata with {len(meta)} entries")
    print(f"Unique files to process: {meta['filename'].nunique()}")
    print("First few entries:")
    print(meta.head())

# ==== 2. Define EMG filter utilities ====
def bandpass_filter(signal, lowcut=20, highcut=450, fs=1000, order=4):
    nyq = 0.5 * fs
    low, high = lowcut/nyq, highcut/nyq
    b, a = butter(order, [low, high], btype="band")
    return filtfilt(b, a, signal, axis=0)

def preprocess_emg(emg, fs=1000):
    emg_filtered = bandpass_filter(emg, fs=fs)
    emg_rect = np.abs(emg_filtered)
    window = 50
    emg_processed = np.zeros_like(emg_rect)
    for i in range(emg_rect.shape[1]):
        emg_processed[:, i] = np.sqrt(
            np.convolve(emg_rect[:, i]**2, np.ones(window)/window, mode='same')
        )
    emg_normalized = (emg_processed - emg_processed.mean(axis=0)) / (emg_processed.std(axis=0) + 1e-8)
    return emg_normalized

def preprocess_angles(angles):
    return (angles - np.mean(angles, axis=0)) / (np.std(angles, axis=0) + 1e-8)

# ==== 3. Process each file in metadata ====
processed_files = set()
failed_files = []

print("\nStarting file processing...")

for idx, row in meta.iterrows():
    fname = row['filename']
    stage = row['stage']
    start, end = row['start'], row['end']

    print(f"\nProcessing entry {idx+1}/{len(meta)}")
    print(f"File: {fname}")
    print(f"Stage: {stage}, Start: {start}, End: {end}")

    if not os.path.exists(fname):
        print(f"❌ Missing file: {fname}")
        failed_files.append(fname)
        continue

    try:
        with h5py.File(fname, 'r') as f:
            time = f['emg2pose/timeseries']['time'][:]
            emg = f['emg2pose/timeseries']['emg'][:]
            angles = f['emg2pose/timeseries']['joint_angles'][:]

            print(f"📊 Data shapes - Time: {time.shape}, EMG: {emg.shape}, Angles: {angles.shape}")

        mask = (time >= start) & (time <= end)
        time_filtered = time[mask]
        emg_filtered = emg[mask]
        angles_filtered = angles[mask]

        if len(time_filtered) == 0:
            print(f"⚠️ No data in range [{start}, {end}] for {fname}")
            continue

        emg_processed = preprocess_emg(emg_filtered)
        angles_processed = preprocess_angles(angles_filtered)

        df_segment = pd.DataFrame()
        for i in range(emg_processed.shape[1]):
            df_segment[f'emg_{i}'] = emg_processed[:, i]
        for j in range(angles_processed.shape[1]):
            df_segment[f'angle_{j}'] = angles_processed[:, j]

        df_segment['time'] = time_filtered
        df_segment['stage'] = stage
        df_segment['filename'] = fname
        df_segment['session'] = row.get('session', 'unknown')
        df_segment['user'] = row.get('user', 'unknown')
        df_segment['side'] = row.get('side', 'unknown')

        # ==== Save each file individually ====
        csv_name = f"processed_{os.path.splitext(os.path.basename(fname))[0]}.csv"
        df_segment.to_csv(csv_name, index=False)
        print(f"💾 Saved processed file: {csv_name}")

        processed_files.add(fname)

    except Exception as e:
        print(f"❌ Error processing {fname}: {str(e)}")
        failed_files.append(fname)
        continue

# ==== 4. Save processing summary ====
summary_file = "processing_summary.csv"
summary = {
    'total_entries_in_metadata': len(meta),
    'unique_files_processed': len(processed_files),
    'failed_files_count': len(failed_files),
    'failed_files': failed_files
}
summary_df = pd.DataFrame([summary])
summary_df.to_csv(summary_file, index=False)
print(f"\n📋 Processing summary saved as: {summary_file}")

# ==== 5. Save automatically to Google Drive ====
from google.colab import drive
drive.mount('/content/drive')

# Path inside Google Drive
save_dir = "/content/drive/MyDrive/EMG_Project"  # <-- you can rename this folder
os.makedirs(save_dir, exist_ok=True)

output_file = os.path.join(save_dir, "prepare_alldataset.csv")
master_df.to_csv(output_file, index=False)
print(f"💾 Master dataset automatically saved as: {output_file}")

# Save processing summary
summary_file = os.path.join(save_dir, "prepare_alldataset_summary.csv")
summary = {
    'total_entries_in_metadata': len(meta),
    'successful_segments_processed': len(all_data),
    'unique_files_processed': len(processed_files),
    'failed_files_count': len(failed_files),
    'final_dataset_shape': master_df.shape,
    'stages': list(master_df['stage'].unique()),
    'users': list(master_df['user'].unique()),
    'sides': list(master_df['side'].unique())
}

summary_df = pd.DataFrame([summary])
summary_df.to_csv(summary_file, index=False)
print(f"📋 Processing summary automatically saved as: {summary_file}")
```

```python
# =====================================
# STEP 1: Setup & Import Libraries
# =====================================
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, Subset
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Path to dataset in Drive (change filename)
csv_path = "/content/drive/MyDrive/EMG_Project/prepare_alldataset.csv"
```

```python

# =====================================
# STEP 2: Custom Dataset with Sliding Windows
# =====================================
class EMGDataset(Dataset):
    def __init__(self, file_path, seq_len=200):
        self.data = pd.read_csv(file_path)
        self.seq_len = seq_len

        # Features (16 EMG channels)
        self.emg = self.data[[f"emg_{i}" for i in range(16)]].values.astype(np.float32)

        # Labels
        self.angles = self.data[[f"angle_{i}" for i in range(15)]].values.astype(np.float32)
        self.stages, _ = pd.factorize(self.data["stage"])  # categorical encoding

    def __len__(self):
        return len(self.emg) - self.seq_len

    def __getitem__(self, idx):
        X = self.emg[idx:idx+self.seq_len]
        y_angles = self.angles[idx+self.seq_len-1]
        y_stage = self.stages[idx+self.seq_len-1]
        return torch.tensor(X), torch.tensor(y_angles), torch.tensor(y_stage)
```

```python

# =====================================
# STEP 3: DataLoader with Train/Val Split
# =====================================
def get_dataloaders(csv_path, seq_len=200, batch_size=64, val_split=0.2):
    dataset = EMGDataset(csv_path, seq_len)
    train_idx, val_idx = train_test_split(np.arange(len(dataset)), test_size=val_split, shuffle=True)

    train_subset = Subset(dataset, train_idx)
    val_subset = Subset(dataset, val_idx)

    train_loader = DataLoader(train_subset, batch_size=batch_size, shuffle=True, num_workers=2)
    val_loader = DataLoader(val_subset, batch_size=batch_size, shuffle=False, num_workers=2)

    return train_loader, val_loader, len(np.unique(dataset.stages))
```

```python
# =====================================
# STEP 4: Define LSTM Model
# =====================================
class LSTMModel(nn.Module):
    def __init__(self, input_dim=16, hidden_dim=128, num_layers=2, num_angles=15, num_stages=15):
        super(LSTMModel, self).__init__()
        self.lstm = nn.LSTM(input_dim, hidden_dim, num_layers, batch_first=True, dropout=0.3)
        self.fc_angles = nn.Linear(hidden_dim, num_angles)   # regression
        self.fc_stage = nn.Linear(hidden_dim, num_stages)    # classification

    def forward(self, x):
        _, (h, _) = self.lstm(x)
        h = h[-1]  # last layer hidden state
        return self.fc_angles(h), self.fc_stage(h)
```

```python


# =====================================
# STEP 5: Training + Validation Loop
# =====================================
def train_model(train_loader, val_loader, num_stages, num_epochs=20, lr=1e-3, device="cuda"):
    model = LSTMModel(num_stages=num_stages).to(device)
    criterion_angles = nn.MSELoss()
    criterion_stage = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)

    for epoch in range(num_epochs):
        # -------- Training --------
        model.train()
        train_loss = 0
        for X, y_angles, y_stage in train_loader:
            X, y_angles, y_stage = X.to(device), y_angles.to(device), y_stage.to(device)

            optimizer.zero_grad()
            pred_angles, pred_stage = model(X)

            loss = criterion_angles(pred_angles, y_angles) + criterion_stage(pred_stage, y_stage)
            loss.backward()
            optimizer.step()
            train_loss += loss.item()

        # -------- Validation --------
        model.eval()
        val_loss, correct, total = 0, 0, 0
        with torch.no_grad():
            for X, y_angles, y_stage in val_loader:
                X, y_angles, y_stage = X.to(device), y_angles.to(device), y_stage.to(device)
                pred_angles, pred_stage = model(X)

                loss = criterion_angles(pred_angles, y_angles) + criterion_stage(pred_stage, y_stage)
                val_loss += loss.item()

                _, predicted = torch.max(pred_stage, 1)
                correct += (predicted == y_stage).sum().item()
                total += y_stage.size(0)

        acc = correct / total
        print(f"Epoch {epoch+1}: Train Loss={train_loss/len(train_loader):.4f}, "
              f"Val Loss={val_loss/len(val_loader):.4f}, Val Acc={acc:.4f}")

    return model
```
